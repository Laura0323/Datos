En primer lugar, se importa la libreria pandas para la lectura del dataframe

```python
import pandas as pd
```

Luego, se lee el archivo CSV con pd.read_csv y se muestra 

```python
df = pd.read_csv("ColombianosRegistradosEnElExterior.csv")
df
```

Despúes se hace un for para determinar en cada iteración se accede a la columna y se llama al método .unique() para determinar los valores únicos de cada columna y luego con f-string para incluir el nombre de la columna y los valores únicos en el mensaje impreso. 

```python
for column in df.columns:
    unique_values = df[column].unique()
    print(f"Valores únicos en la columna {column} son: {unique_values}")
```

Posteriormente se elimina la columna 'Localización' con el drop y el axis=1 se usa para especificar que se va a eliminar una columna y se toma esta desición ya que no aporta información relevante para lo que se quiere analizar. 

```python
df = df.drop('Localización', axis=1)
df
```

Luego, se eliminan los registros con estura menor a 150 cm y mayor a 200 cm, filtrando solo los valores que si se encuentran dentro de ese rango. 

```python
df = df[(df['Estatura (CM)'] >= 150) & (df['Estatura (CM)'] <= 200)]
df
```

Despúes, se utilizar el operador ~ para negar una condición boolenana, por lo tanto, se toma la columna 'Área conocimiento' y con el str.contains se comprueba que tenga el texto 'NO INDICA' y 'NO REGISTRA', es decir, el operador ~ devuelve las filas que no cumplen con esas condiciones. 

```python
df = df[~(df['Área Conocimiento'].str.contains('NO INDICA')) & ~(df['Área Conocimiento'].str.contains('NO REGISTRA'))]
df
```

Posteriormente haciendo uso del mismo operador lógico ~ se toma la columna 'Nivel académico' y con el str.contains valida que contenga el texto 'NO INDICA' y 'NO REGISTRA', eliminando todas las filas que coincidan con ambas condiciones

```python
df = df[~(df['Nivel Académico'].str.contains('NO INDICA')) & ~(df['Nivel Académico'].str.contains('NO REGISTRA'))]
df
```

También, utilizando el mismo operador ~ se valida que la columna 'Estado civil' contenga el texto 'DESCONOCIDO' para eliminar todas las filas que contengan ese valor 

```python
df = df[~df['Estado civil'].str.contains('DESCONOCIDO')]
df
```

Con el operador ~ se verifica que la columna 'Género' con el str.contains contenga el texto 'DESCONOCIDO' y elimine las filas que contengan dicho dato

```python
df = df[~df['Género'].str.contains('DESCONOCIDO')]
df
```

Se repite el procedimiento anterior pero esta vez tomando la columna 'Etnia de la persona' y con el str.contains se valida que contenga el texto 'SIN ETNIA REGISTRADA' y por consiguiente elimina los registros que coincidan con este campo

```python
df = df[~df['Etnia de la persona'].str.contains('SIN ETNIA REGISTRADA')]
df
```

Posteriormente, se hace una copia del dataframe, y a dicha copia se le reemplaza específicamente Área Conocimiento AGRONOMÍA, VETERINARIA Y AFINES por AGRONOMÍA, VETERINARIA Y ZOOTECNIA con el método .replace() y luego se le reasigna la copia al data frame original

```python
df_copy = df.copy()

df_copy['Área Conocimiento'] = df_copy['Área Conocimiento'].replace('AGRONOMÍA, VETERINARIA Y AFINES', 'AGRONOMÍA, VETERINARIA Y ZOOTECNIA')

df = df_copy
df
```

Finalmente, lo que se hizo fue hacer un conteo para saber el estado civil de colombianos en el exterior, para ello se hizo uso del str.strip() y str.lower() en el que se convierte el formato de texto en minusculas y se eliminan los espacios para no tener ningun inconveniente con el dataframe que por defecto esta en mayusculas.

Lo mismo se hizo para analizar el área de conocimiento de los colombianos y tener en cuenta qué es lo que mas ejercen laboralmente en otros países.

```python

df['Estado Civil'] = df['Estado civil'].str.strip().str.lower()

conteo_estados_civiles = df['Estado civil'].value_counts()

conteo_estados_civiles
```

```python
df['Área Conocimiento'] = df['Área Conocimiento'].str.strip().str.lower()

conteo_area_de_conocimiento = df['Área Conocimiento'].value_counts()

conteo_area_de_conocimiento
```
