El primer paso es importar las librerias necesarias para realizar los gráficos.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Una vez ejecutado procedemos a cargar el dataframe de tienda_ventas.csv haciendo uso de la librería pandas

```python
df = pd.read_csv('tienda_ventas.csv')
df
```

A continuación se responderá la siguiente pregunta usando el dataframe:

1. ¿Cómo ha cambiado la venta de mis productos por mes y por año?

Esta línea de código convierte la columna "date" al tipo de dato de fecha.

```python
df['date'] = pd.to_datetime(df['date'])
df
```

Se crea la columna 'Año-Mes' utilizando lambda y strftime.

```python
df['Año-Mes']  = df['date'].apply(lambda x: x.strftime('%Y/%m'))
df
```

Se crea una nuevo dataframe llamada 'ventas_por_año_y_mes' en el cual se va a agrupar la nueva columna generada, que en este caso es 'Año-Mes' por medio de la función .groupby, posteriormente se toma la columan 'sales' y se suma la cantidad de dinero generado por las ventas que se hicieron en ese año y mes por medio de la función sum().

```python
ventas_por_año_y_mes = df.groupby(['Año-Mes'])['sales'].sum().reset_index()
ventas_por_año_y_mes
```

 Para la creación del gráfico de línea primero se usa figsize para establecer el tamaño del gráfico, luego se toma la columna 'Año-Mes' y la sumatoria de las ventas 'sales' y el plt.plot recibe dichos parámetros para poder graficar las ventas que se hicieron por cada año, posteriormente se usa el atributo marker para establecer los puntos que se muestran en el gráfico, el linestyle para establecer la línea por la cual se va a unir cada punto y establecer el color que es este caso es el rojo. 
 Después procedemos a personalizar el gráfico para ponerle el nombre del gráfico que en este caso es 'Evolución de las ventas por año y mes', los nombres de los ejes X y Y que es este caso es x para para la nueva columna establecida, que es 'Año-Mes' vs ventas totales realizadas cada año para el eje Y. por último, se utiliza el atributo .xtikcs con una rotación de 45 para que las etiquetas del eje x puedan tener mejor legibilidad y el show finalmente es para mostrar el gráfico. 

```python
plt.figure(figsize=(12, 6))
plt.plot(ventas_por_año_y_mes['Año-Mes'], ventas_por_año_y_mes['sales'], marker='o', linestyle='-', color='r')
plt.xlabel('Año-Mes')
plt.ylabel('Total de ventas')
plt.title('Ventas de productos por Año y Mes')

plt.xticks(rotation=45)  

plt.show()
```

La segunda pregunta para contestar por medio del gráfico de líneas es:
2. En el año 2015, como se comparan las ventas de tres familias de productos diferentes a lo largo del tiempo (SEAFOOD, DAIRY, MEATS)

Filtra los datos utilizando la columna 'Año-Mes' para incluir solo el año 2015 en un nuevo dataframe llamado 'df_2015'. luego, se usa el startswith para solo filtrar las filas con el valor '2015'

```python
df_2015 = df[df['Año-Mes'].str.startswith('2015')]
df_2015
```

Filtra los datos en un nuevo dataframe llamado 'familias' para incluir solo los productos SEAFOOD, DAIRY y MEATS y luego otro dataframe llamado 'df_familias' para filtrar solo estos tres productos pero solo en el año 2015. el método isin devuelve una serie booleana que indica si cada fila cumple o no con la condición especificada, es decir, que las filas que se filtraron en la columna 'familiy' coincidan con los tres productos seleccionados.

```python
familias = ['SEAFOOD', 'DAIRY', 'MEATS']
df_familias = df_2015[df_2015['family'].isin(familias)]
df_familias
```

Agrupa los datos por mes y familia por medio de la función .groupby en un nuevo dataframe llamado 'ventas_por_mes_y_familia' y suma las ventas en cada grupo de productos con la función sum(), y la función .unstack() es usada para desapilar un índice en el dataframe, es decir, toma la columna de 'sales' y la convierte en fila.

```python
ventas_por_mes_y_familia = df_familias.groupby(['Año-Mes', 'family'])['sales'].sum().unstack()
ventas_por_mes_y_familia
```

Para la creación del gráfico de línea con el atributo figsize se establece el tamaño y se hace un ciclo for para el dataframe llamado 'familia' donde se van a almacenar los datos agrupados para crear el gráfico, posteriormente, en 'familias' se encuentran almacenados las 3 familias de productos SEAFOOD, DAIRY Y MEATS, de igual manera, 'ventas_por_mes_y_familia.index' representa las fechas (meses y años) y 'ventas_por_mes_y_familia[familia]' representa las ventas de cada familia de productos. 
Por último, se personaliza el gráfico dando los nombres a ambos ejes, para el X serían las fechas del año en que realizaron las ventas, en este caso solo 2015, y para el eje Y la cantidad de ventas realizadas, se le asignó un título al gráfico y se utilizó el plt.legend() para mostrar la leyenda de la familia de los tres productos. Posteriormente se uso .xticks para rotar las etiquetas del eje Y para tener mejor visibilidad y show para finalmente mostrar el gráfico. 

```python
plt.figure(figsize=(12, 6))

for familia in familias:
    plt.plot(ventas_por_mes_y_familia.index, ventas_por_mes_y_familia[familia], label=familia, marker='o', linestyle='-')

plt.xlabel('Año y Mes')
plt.ylabel('Ventas')
plt.title('Ventas de Productos en el Año 2015')
plt.legend() 

plt.xticks(rotation=45) 

plt.show() 
```
